# Forecast Problem Framing

This notebook records the agreed framing for the dengue forecasting task.

## Problem Definition

- **Task**: Multi-series weekly dengue case forecasting across Bangladesh geographies.
- **Target**: Weekly dengue case counts on the original count scale.
- **Series type**: Panel time series with one row per geography and week.
- **Forecast use case**: Short-term operational forecasting with longer-horizon experimental forecasting.

## Horizon Selection

- **Primary horizon**: 1 to 4 weeks ahead.
- **Secondary horizon**: 5 to 12 weeks ahead.
- **Reasoning**:
  - Short-horizon persistence is strong in the observed series.
  - Longer horizons are still useful for research, but uncertainty and model risk increase quickly.
  - Weather coverage currently stops before the latest case coverage, so short-horizon autoregressive forecasts are the safest starting point.

## Grain Selection

- **Temporal grain**: Weekly.
- **Spatial grain**: `geo_id`.
- **Modeling unit**: One forecast per geography per week.
- **Hierarchy note**:
  - `DHA_DIV` overlaps with `DHA_METRO`.
  - If coherent aggregation is required, use `dhaka_div_out_metro` plus `DHA_METRO` instead of keeping both overlapping nodes.

## KPI Selection

- **Primary KPI**: Mean Absolute Error (MAE), macro-averaged across geography and horizon.
- **Secondary KPI**: Root Mean Squared Error (RMSE) on counts.
- **Why MAE first**:
  - It is easier to interpret on case counts.
  - It is less dominated by the largest outbreak spikes than RMSE.
  - It makes per-geo comparisons more stable across high- and low-incidence series.
- **Why keep RMSE**:
  - It still penalizes missed outbreak peaks, which matter operationally.
- **Future extension**:
  - Add interval coverage and interval width once probabilistic forecasts are introduced.
